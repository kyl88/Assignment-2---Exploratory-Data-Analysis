```python
# find the median age using the interquartile range

print(df.Age.quantile(.50)) # Q2 - median of the Age group
```

```python
# List all the students who are not in High School

df[(df['Age'] > 18)]
```

```python
# find how many students did not finish high school and selected the Product Design category
filt=df['Age'] == 16
df.loc[filt]['Virtual Internship Group Category'].str.contains('Product Design').sum()
```

```python
# find how many students who graduated at 23 years old and selected the Product Development category
filt=df['Age'] == 23
df.loc[filt]['Virtual Internship Group Category'].str.contains('Product Development').sum()
```

```python
# find how many students are aged at 45 years old and selected the Product Development category
filt=df['Age'] == 45
df.loc[filt]['Virtual Internship Group Category'].str.contains('Product Marketing').sum()
```

```python
# Find out how many students are middle-aged and selected the Product Development category

filt=df['Age'] == 45
df.loc[filt]['Virtual Internship Group Category'].str.contains('Product Marketing').sum()
```

```python
# find all the students who are still in high school and recently and reside in the USA who are in the Radical X internship

df[(df.Age <=18) & (df.Country == 'USA')]
```
